```python
#!/usr/bin/env python3
import random
from tabulate import *

def initialize():
    print('Welcome to the genetic mapping simulator.')
    print('What type of genetic cross would you like to observe?')
    print('Option 1: 3-1 cross')
    print('Press 1 for Option 1')
    print('or')
    print('Option 2: 2-2 cross')
    print('Press 2 for Option 2')
    crossType = int(input())

    print('Please choose a sample size larger than 100: ')
    sampleSize = int(input('Sample Size # = '))    

    print('Would you like to use test data?')
    randomize = input('Y or N? \n')
    #output will return 1 or 2 with the tuple of map distances. 1/2 will indicate the type of cross 
    data = []
    if randomize == 'Y' or randomize == 'y':
        A = random.randint(0,13)
        B = random.randint(14,26)
        C = random.randint(27,40)
        D = random.randint(40,50)
        #interference = random.uniform(0,1)
        data = [A, B, C, D]
    else:
        A = int(input('A: '))
        B = int(input('B: '))
        C = int(input('C: '))
        D = int(input('D: '))
        #interference = int(input('interference: '))
        data = [A, B, C, D]
    return (crossType, data, sampleSize)
```

```python
def mapDistChr1(chromosome1):
    '''[Matt] input chromosome 1 to calculate the Map Units between each gene. Supports 2 or 3 genes'''
    if len(chromosome1) == 3:
        #3-1 cross
        AB = abs(chromosome1[0] - chromosome1[1])
        BC = abs(chromosome1[1] - chromosome1[2])
        return [AB, BC]
    elif len(chromosome1) == 2:
        #2-2 cross
        AB = abs(chromosome1[0] - chromosome1[1])
        return AB

def progenyNum(recombFrequency, N):
    kids = []
    for n in recombFrequency:
        kids.append(int(n*N))
    nonRecombinant = (.5 * N) - sum(kids)
    kids.append(nonRecombinant)
    return kids

#in execute, if initialize[0] == 1, call this function
def cross31(data, sampleSize): #take in second item (the list) from the initialize function
    A = data[0]
    B = data[1]
    C = data[2]
    D = data[3]

    N = sampleSize

    chromosome1 = [A, B, C]
    chromosome2 = D

    mapDist1 = mapDistChr1(chromosome1)
    mapDist2 = 0

    AB = mapDist1[0]
    BC = mapDist1[1]
    recomb1 = (AB / 100) * (1 - (BC / 100))
    recomb2 = (BC / 100) * (1 - (AB / 100))
    recomb3 = (AB / 100) * (BC / 100)

    recombinants = [ recomb1, recomb2, recomb3 ]

    prog = progenyNum(recombinants, N)

    case1 = tabulate(
        [
            ['ABC', 'nonrecombinant', prog[3]], 
            ['abc','nonrecombinant', prog[3]],
            ['aBC','single crossover AB',prog[0]],
            ['Abc','single crossover AB',prog[0]],
            ['ABc','single crossover BC',prog[1]],
            ['abC','single crossover BC',prog[1]],
            ['AbC','double crossover',prog[2]],
            ['aBc','double crossover',prog[2]]
            ], 
            headers=['Genotype', 'Type','Number of Progeny'])

    return case1

def cross22(data, sampleSize):
    A = data[0]
    B = data[1]
    C = data[2]
    D = data[3]

    N = sampleSize

    chromosome1 = [A, B]
    chromosome2 = [C, D]

    mapDist1 = mapDistChr1(chromosome1)
    mapDist2 = abs(chromosome2[0] - chromosome2[1])

    AB = mapDist1
    CD = mapDist2

    CDcross = (CD / 100) * (1 - (AB / 100))
    ABcross = (AB / 100) * (1 - (CD / 100))
    ABCDcross = (AB / 100) * (CD / 100)

    recombinants = [CDcross, ABcross, ABCDcross]
    
    prog = []
    for n in recombinants:
        prog.append(int(.5*n*N))
    nonRecombinant = (.25 * N) - sum(prog)
    prog.append(nonRecombinant)

    case2 = tabulate(
        [
            ['AB-CD', 'nonrecombinant', prog[3]], 
            ['AB-cd','nonrecombinant', prog[3]],
            ['ab-cd','nonrecombinant',prog[3]],
            ['ab-cD','nonrecombinant',prog[3]],
            ['AB-Cd','CD crossover', prog[0]],
            ['AB-cD','CD crossover', prog[0]],
            ['ab-Cd','CD crossover', prog[0]],
            ['ab-cD','CD crossover', prog[0]],
            ['Ab-CD','AB crossover', prog[1]],
            ['Ab-cd','AB crossover', prog[1]],
            ['aB-CD','AB crossover', prog[1]],
            ['aB-cd','AB crossover', prog[1]],
            ['Ab-Cd','AB/CD crossover', prog[2]],
            ['Ab-cD','AB/CD crossover', prog[2]],
            ['aB-Cd','AB/CD crossover', prog[2]],
            ['aB-cD','AB/CD crossover', prog[2]]
            ], 
            headers=['Genotype', 'Type','Number of Progeny'])
    return case2
```

```python
def execute():

    init = initialize()

    sampleSize = init[2]
    data = init[1]
    crossType = init[0]

    if crossType == 1:
        print(cross31(data, sampleSize))

    elif crossType == 2:
        print(cross22(data, sampleSize))

execute()
```
